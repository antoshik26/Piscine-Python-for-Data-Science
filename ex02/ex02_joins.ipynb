```python
df = pd.read_sql('select t.uid, t.labname, t.timestamp as first_commit_ts, d.datetime as first_view_ts from checker t left JOIN pageviews d on t.uid=d.uid where t.status = "ready" and t.numTrials = "1" and t.labname in ("laba04", "laba04s", "laba05", "laba06", "laba06s", "project1") and t.uid like "user%" and (d.datetime is null or d.datetime=(select min(pageviews.datetime) from pageviews where uid = t.uid))', connect_sql, parse_dates=['first_view_ts', 'first_commit_ts'])
```
