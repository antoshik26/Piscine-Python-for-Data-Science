```python
df = pd.read_csv('/content/drive/MyDrive/auto.csv', index_col = 'ID', engine='python')
df
```

```python
df.count()
```

```python
df = df.drop_duplicates(subset = ['CarNumber', 'Make_n_model', 'Fines'], keep = 'last')
```

```python
df.count()
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis=1, thresh=500)
df
```

```python
df.isna().sum()
```

```python
df['Refund'] = df.Refund.fillna(method ='ffill')
df
```

```python
df.isna().sum()
```

```python
mean_fine = df['Fines'].mean(skipna=True)
df['Fines'] = df['Fines'].fillna(mean_fine)
df
```

```python
df.isna().sum()
```

```python
a = df['Make_n_model']
df['Make'] = df.apply(lambda x: x['Make_n_model'].split(' ')[0], axis=1)
df['Make_n_model'] = a
df['Model'] = df.apply(lambda x: x['Make_n_model'].split(' ')[1] if len(x['Make_n_model'].split()) > 1 else pd.isnull(x), axis=1, result_type='expand')
df
```

```python
df.to_json('/content/drive/MyDrive/auto.json', orient='records')
```
